```python
import pandas as pd
import re


# Read stop words from the text file
with open('stopwords.txt', 'r') as file:
    stop_words = file.read().splitlines()

# Function to clean and preprocess text
def preprocess_text(text):
    # Convert float to string
    text = str(text)
    # Remove extra whitespaces and newlines
    text = re.sub(r'\s+', ' ', text)
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenize the text
    tokens = text.split()
    # Remove stop words
    tokens = [word for word in tokens if word.lower() not in stop_words]
    # Join tokens back into a single string
    preprocessed_text = ' '.join(tokens)
    return preprocessed_text

# Apply cleaning and preprocessing function to the 'SECTION_TEXT' column
df2['SECTION_TEXT'] = df2['SECTION_TEXT'].astype(str).apply(preprocess_text)

# Save the preprocessed data back into the same column
df2.to_csv('preprocessed_dataset.csv',index=False)
```

# Vocabulary

```python
# Initialize an empty dictionary to store the vocabulary
vocabulary = {}

# Initialize a counter to assign unique IDs to terms
term_id_counter = 0

# Iterate over each row in the 'SECTION_TEXT' column
for text in df2['SECTION_TEXT']:
    # Tokenize the text
    tokens = text.split()
    
    # Iterate over each token in the text
    for token in tokens:
        # Check if the token is already in the vocabulary
        if token not in vocabulary:
            # If not, assign a unique ID to the token and update the vocabulary
            vocabulary[token] = term_id_counter
            term_id_counter += 1

# Print the vocabulary
print("Vocabulary:")
print(vocabulary)
```

# Term Frequency

```python
import pandas as pd

# Sample DataFrame
# df1 = ...  # Assuming you already have df1 with the 'SECTION_TEXT' column

# Initialize an empty list to store TF representations
tf_representations = []

# Iterate over each row in the DataFrame
for index, row in df2.iterrows():
    text = row['SECTION_TEXT']
    
    # Tokenize the text
    tokens = text.split()
    
    # Count the frequency of each term
    term_freq = {}
    for token in tokens:
        term_id = vocabulary.get(token)  # Get the term ID from the vocabulary
        if term_id is not None:
            term_freq[term_id] = term_freq.get(term_id, 0) + 1
    
    # Create tuples for each term containing its ID and frequency
    tf_representation = [(term_id, freq) for term_id, freq in term_freq.items()]
    
    # Append the TF representation to the list
    tf_representations.append(tf_representation)

# Add the TF representations to the DataFrame as a new column
df2['TF'] = tf_representations

# Print the updated DataFrame
df2
```

# Removing NonZeros Values

```python
import pandas as pd

# Sample DataFrame
# df1 = ...  # Assuming you already have df1 with the 'SECTION_TEXT' and 'TF' columns

# Initialize an empty list to store sparse TF representations
sparse_tf_representations = []

# Iterate over each row in the DataFrame
for index, row in df2.iterrows():
    tf_representation = row['TF']
    
    # Filter out zero-frequency terms and store non-zero entries
    sparse_tf_representation = [(term_id, freq) for term_id, freq in tf_representation if freq != 0]
    
    # Append the sparse TF representation to the list
    sparse_tf_representations.append(sparse_tf_representation)

# Update the 'TF' column with the sparse TF representations
df2['TF'] = sparse_tf_representations

# Print the updated DataFrame
df2
```

```python
from mrjob.job import MRJob
import pandas as pd

class VocabularyBuilder(MRJob):
    
    def mapper(self, _, line):
        # Split the line into article_id and section_text
        article_id, section_text = line.split('\t')
        
        # Tokenize the section_text
        tokens = section_text.split()
        
        # Emit each token as key with a count of 1
        for token in tokens:
            yield token, 1

    def reducer(self, token, counts):
        # Sum up the counts for each token to get its total frequency
        total_count = sum(counts)
        
        # Emit the token and its total frequency
        yield token, total_count

if __name__ == '_main_':
    # Read the DataFrame
    df2 = pd.read_csv("preprocessed_dataset1.csv")  # Provide the path to your DataFrame
    
    # Create an instance of the MapReduce job
    mr_job = VocabularyBuilder(args=['-r', 'local'])
    
    # Provide input data using stdin
    with mr_job.make_runner() as runner:
        # Concatenate article_id and section_text with tab separation and feed into stdin
        runner.run(stdin='\n'.join(f"{article_id}\t{section_text}" for article_id, section_text in zip(df2['ARTICLE_ID'], df2['SECTION_TEXT'])))
        
        # Parse the output and build the vocabulary
        vocabulary = {}
        for line in runner.stream_output():
            token, total_count = mr_job.parse_output_line(line)
            vocabulary[token] = total_count
    
    # Print the vocabulary
    print("Vocabulary:")
    print(vocabulary)
```
